````python
#10. Пользователь поочередно вводит координаты точки в декартовой системе 
#координат. Определить, какой четверти принадлежит данная точка или на какой 
#оси она находится. Расположение точки вывести на экран. Найти произведение 
#номера четверти на расстояние от этой точки до начала координат и вывести его на экран. 
#Если точка лежит на оси, считать, что номер четверти равен 0.
import math
while True:
    x = input("Координата x (или 'x' для выхода):  ")
    if x == 'x':
        break

    y = input("Координата y (или 'x' для выхода):  ")
    if y == 'x':
        break

    try:
        x = float(x)
        y = float(y)

        if x == 0 or y == 0: ch = 0
        elif x > 0:
            if y > 0:
                ch = 1
            else:
                ch = 4
        else:
            if y > 0:
                ch = 2
            else:
                ch = 3

        S = math.sqrt(x**2 + y**2)

        prod = ch * S

        print(f"Точка ({x}, {y}) находится в четверти {ch}.")
        print(f"Произведение четверти и расстояния: {prod:.2f}\n")

    except:
        print("Неверный ввод данных")
````

````python
#11. Выведите значение заданного целого числа от 0 до 999 прописью. 
#Например, «сто девяносто один» для числа 191, «одиннадцать» для числа 11. 

def words(n):
    ed = ["", "один", "два", "три", "четыре", "пять", "шесть", "семь", "восемь", "девять"]
    des = ["", "одиннадцать", "двенадцать", "тринадцать", "четырнадцать", "пятнадцать", "шестнадцать", "семнадцать", "восемнадцать", "девятнадцать"]
    eddes = ["", "десять", "двадцать", "тридцать", "сорок", "пятьдесят", "шестьдесят", "семьдесят", "восемьдесят", "девяносто"]
    sot = ["", "сто", "двести", "триста", "четыреста", "пятьсот", "шестьсот", "семьсот", "восемьсот", "девятьсот"]

    if 0 <= n <= 9:
        return ed[n]
    elif 11 <= n <= 19:
        return des[n - 10]
    elif 21 <= n <= 99:
        return eddes[n // 10] + (" " + ed[n % 10] if n % 10 != 0 else "")
    elif 101 <= n <= 999:
        return sot[n // 100] + (" " + words(n % 100) if n % 100 != 0 else "")
    else:
        return "Число вне диапазона (0-999)"


try:
    n = int(input("Введите целое число от 0 до 999: "))
    if 0 <= n <= 999:
        words = words(n)
        print(f"{n}: {words}")
    else:
        print("Число вне диапазона (0-999)")
except ValueError:
    print("Некорректный ввод. Пожалуйста, введите целое число от 0 до 999.")

print("Программа завершена.")

````

````python
#12. Вычислите значение выражения, которое состоит из целых чисел и знаков «+» и «-». 
#Выражение вводится как символьная строка.

def calc(tokens):
        stack = []
        current_num = 0
        operator = 1 

        i = 0
        while i < len(tokens):
            token = tokens[i]
            if token == "+":
                operator = 1
            elif token == "-":
                operator = -1
            else:
                current_num = 0
                while i < len(tokens) and tokens[i].isdigit():
                    current_num = current_num * 10 + int(tokens[i])
                    i += 1
                
                stack.append(operator * current_num)
                continue
            i += 1

        
        result = sum(stack)
        return result

str = input()
str = str.replace(" ", "")
result = calc(list(str))
print(f"Результат: {result}")

````

````python
#14. Реализовать калькулятор, который принимает от пользователя 
#через пробел строку следующего вида: «a op b», где a и b – некоторые числа, 
#а ‘op’ определяет оператор и может принимать значения $«+,-,*,/,**,%»$. 
#В зависимости от оператора с помощью форматирования строк вывести результат 
#в виде: «a + b = 3 + 2 = 5», где была получена строка «3 + 2». 
#Для операций возведения в степень и деления по модулю использовать вместо 
#знака оператора соответствующие выражения. Например, для строки «2 ** 3» 
#должно быть выведено «a в степени b = 2 в степени 3 = 8».

str = input()

try:
    a, op, b = str.split()
    a = int(a)
    b = int(b)
    res = None
    oper = ''

    if op == '+':
        res = a + b
        oper = '+'
    elif op == '-':
        res = a - b
        oper = '-'
    elif op == '*':
        res = a * b
        oper = '*'
    elif op == '/':
        if b == 0:
            print("Ошибка: Деление на ноль")
        res = a / b
        oper = '/'
    elif op == '**':
        res = a ** b
        oper = 'в степени'
    else:
        print("Ошибка: Неверный оператор. Поддерживаемые операторы: +, -, *, /, **.")

    print (f"a {oper} b = {a} {oper} {b} = {res}")

except:
    print("Неверный ввод данных")
````

````python
#16. Имеются два n – мерных вектора x и y, 
#которые задают координаты n точек на плоскости 
#(случайные целые числа). Найти наиболее близкие друг другу точки.

from math import inf, dist
from random import randint
 
n = int(input('n = '))
x = [randint(1, n * 2) for _ in range(n)]
y = [randint(1, n * 2) for _ in range(n)]
print(x)
print(y)
p = list(zip(x,y))
min_dist = inf
for i in range(n):
    for j in range(i + 1, n):
        min_dist = min(min_dist, dist(p[i], p[j]))
print(min_dist)
````
